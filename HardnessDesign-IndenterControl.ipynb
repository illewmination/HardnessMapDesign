```python
#Average Sample Image
image_list = ['1', '2', '3', '4', '5', '6', '7']
for image_name in image_list:    
    #OG Image
    path_to_sample="./Control_Revisions/"+image_name+".png" #Human_Test_Map.png"
    img = image.load_img(path_to_sample)
    magfactor = 1#2
    width = np.shape(img)[0]
    height = np.shape(img)[1]
    newwidth = 200#int(magfactor * width)
    newheight = 200#int(magfactor * height)
    img = img.resize((newheight, newwidth), Image.ANTIALIAS)
    img = image.img_to_array(img)
    img = img/255
    testing_window = img
    testing_window = np.expand_dims(testing_window, axis = 0)
    hardness_value = float(model.predict(testing_window, verbose=0))
    print(hardness_value)
    fig, ax = plt.subplots(figsize=(width/30,height/30))
    pos = ax.imshow(img, cmap='jet', vmin=2.5, vmax=4.8) 
    ax.set_aspect('equal')
    plt.axis('off')
#     fig.colorbar(pos, ax=ax, shrink=0.27)  
#     print(img[:,:,0].shape)    
    
    
    #Avg Image
    avg_r = np.mean(img[:,:,0])
    avg_g = np.mean(img[:,:,1])
    avg_b = np.mean(img[:,:,2])
#     print(avg_r, avg_b, avg_b)
    
    r_val = np.zeros((200,200))+avg_r
    g_val = np.zeros((200,200))+avg_g
    b_val = np.zeros((200,200))+avg_b
#     for i in img[:,:,0]:
#         i = avg_r
#     for i in img[:,:,1]:
#         i = avg_g
#     for i in img[:,:,2]:
#         i = avg_b
    avg_img = np.asarray([r_val, g_val, b_val])    
    avg_img = np.transpose(avg_img,(1,2,0))
#     print(avg_img.shape)
    fig, ax = plt.subplots(figsize=(width/30,height/30))
    pos = ax.imshow(avg_img, cmap='jet', vmin=2.5, vmax=4.8) 
    ax.set_aspect('equal')
    plt.axis('off')
#     fig.colorbar(pos, ax=ax, shrink=0.27)  
    testing_window = avg_img
    testing_window = np.expand_dims(testing_window, axis = 0)
    avg_hardness_value = float(model.predict(testing_window, verbose=0))
    print(avg_hardness_value)
```
